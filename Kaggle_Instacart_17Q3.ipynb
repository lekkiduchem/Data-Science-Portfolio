```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
orders_details = pd.concat([pd.read_csv('data/order_products__train.csv'), 
                            pd.read_csv('data/order_products__prior.csv')]).sort_values(by='order_id')
products = pd.read_csv('data/products.csv')
aisles = pd.read_csv('data/aisles.csv')
departments = pd.read_csv('data/departments.csv')

orders = pd.read_csv('data/orders.csv')
sample_subs = pd.read_csv('data/sample_submission.csv')

# prior = pd.read_csv('data/order_products__prior.csv')
# train = pd.read_csv('data/order_products__train.csv')
```

```python
orders_full_info = pd.merge(left = orders_details, 
                            left_on = 'product_id', 
                            right_on = 'product_id', 
                            how = 'left',
                            right = pd.merge(left = products,
                                             right = departments,
                                             left_on = 'department_id', 
                                             right_on = 'department_id', 
                                             how='left').merge(right = aisles,
                                                               left_on = 'aisle_id', 
                                                               right_on = 'aisle_id', 
                                                               how='left'))

orders_full_info = pd.merge(left=orders_full_info, 
                            right=orders, 
                            left_on='order_id', 
                            right_on='order_id', 
                            how='left').drop(['aisle_id', 'department_id', 'eval_set'], axis=1).sort_values(by='order_id')
```
